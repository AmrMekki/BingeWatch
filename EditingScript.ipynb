```python
import pandas as pd 
```

```python

data = pd.read_csv("MovieGenre.csv",encoding='latin1')
data
```

```python
data['Title']
```

```python
year = []
title = []
```

```python
for i in range(len(data)):
    temp = data['Title'][i] #take the title with year
    temp = temp.split("(") #split into 0:title  1: year)
    title.append(temp[0][:-1]) #add title to title array without final space
    try:
        temp[1] = temp[1][:-1] #remove ) from year
    except:
        year.append('2022') #if no year we insert 2022 temporarily
        continue
    year.append(temp[1])
    
```

```python
titleDF = pd.DataFrame(title)
titleDF
```

```python
n=0
for i in range(len(year)):
    if(year[i]=='2022'):
        n+=1
n #we have 505 record that will be deleted
```

```python
yearDF = pd.DataFrame(year)
yearDF
```

```python
data['Title'] = titleDF
data['Year'] = yearDF
```

```python
data = data[data['Year'] !='2022'] #now we don't have movies that had no year
data
```

```python
data = data.reset_index(drop=True) #reassign index because of deleted rows
data
```
