```python
import pandas as pd
data = pd.read_csv("MultiGenre.csv")
data
```

```python
import os
```

```python
genre = []
imdbID = []
for i in range(len(data)):
    if os.path.exists(f"Posters100x100/{data['imdbID'][i]}.jpg"):
        imdbID.append(data['imdbID'][i])
        genre.append(data['Genre'][i])
print(len(imdbID))
print(len(genre))
    
```

```python
idDF = pd.DataFrame(imdbID)
idDF
```

```python
genreDF = pd.DataFrame(genre)
genreDF
```

```python
newdata = pd.DataFrame()
newdata['imdbID'] = idDF
newdata['Genre'] = genreDF
newdata
```

```python
newdata.to_csv('MultiGenre.csv')
```
