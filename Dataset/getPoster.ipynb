```python
import pandas as pd 
import requests
from keras.models import Sequential
from tensorflow.keras import layers, models
from keras.layers.core import Dense, Dropout
from PIL import Image
import os
import PIL
import glob
```

```python
data = pd.read_csv("../MovieGenreEdited.csv",encoding='latin1')
data
```

```python
for i in range (len(data)) :

    posterUrl = data['Poster'][i]

    img = requests.get(posterUrl, allow_redirects=True)


    open(f"{data['imdbId'][i]}.jpg", 'wb').write(img.content)
```

```python
# data augmentation

image = Image.open(f"{data['imdbId'][0]}.jpg")
print(image.size)


resized_image = image.resize((100,100))
print(resized_image.size)
resized_image.save(f"{data['imdbId'][0]}.jpg")
```

```python
model = models.Sequential()
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
```
