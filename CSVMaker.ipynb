```python
import pandas as pd
from sklearn import preprocessing
```

```python
genreEncoder = preprocessing.LabelEncoder()
genreEncoder.fit(data['Genre'])
data['GenreCategory']= genreEncoder.transform(data['Genre'])
data
```

```python
ourGenre = 'Action'
```

```python
genreCode = 0
for i in range(len(data)):
    if data['Genre'][i] == ourGenre:
        genreCode = data['GenreCategory'][i]
        break
genreCode
```

Count how many of this genre

```python
genreSize = 0
for i in range(len(data)):
    if data['GenreCategory'][i] == genreCode:
        genreSize+=1
genreSize
```

now we make an array of comedy and non comedy (genreSize for each)

```python
arrayOfCSV = []
for i in range(len(data)):
    if data['GenreCategory'][i] == genreCode:
        catNo =  1
    else:
        catNo = 0 
    temp = [data['imdbID'][i], catNo]
    arrayOfCSV.append(temp)
len(arrayOfCSV)
```

now we delete indexes that have same title as comedy

```python
#first make an array of all genre titles
genreTitles = []
for i in range(len(arrayOfCSV)):
    if arrayOfCSV[i][1] == 1:
        genreTitles.append(arrayOfCSV[i][0])
print(len(genreTitles))
print(len(arrayOfCSV))
```

```python
for i in range(len(arrayOfCSV) - 1, -1 , -1):
    if arrayOfCSV[i][1] ==0 and (arrayOfCSV[i][0] in genreTitles):
        arrayOfCSV.pop(i)
len(arrayOfCSV)
```

```python
for i in range(len(arrayOfCSV)-1,-1,-1):
    if arrayOfCSV[i][1] == 0 and len(arrayOfCSV) > (genreSize *2):
        arrayOfCSV.pop(i)
len(arrayOfCSV)
```

```python
genreCSV = pd.DataFrame(arrayOfCSV)
genreCSV = genreCSV.rename({0:'imdbID', 1:'Genre'}, axis = 1)
genreCSV
```

```python
genreCSV.to_csv(f'GenresCSV/{ourGenre}.csv')
```
